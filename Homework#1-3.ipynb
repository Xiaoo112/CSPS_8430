```python
class model(nn.Module):
    def __init__(self, ):
        super().__init__()
        self.hide1 = nn.Linear(784, 250)
        self.hide2 = nn.Linear(250, 125)
        self.out = nn.Linear(125, 10)
    
    def forward(self, x):
        x = nn.functional.leaky_relu(self.hide1(x))
        x = nn.functional.leaky_relu(self.hide2(x))
        x = self.out(x)
        return x
```

```python
def test(model, data_loader):
    loss_fun = nn.CrossEntropyLoss()
    test_epoch = 0.0
    for images, labels in data_loader:
        images, labels = images.to(device).reshape(-1, 28 * 28), labels.to(device)
        predict = model(images)
        loss = loss_fun(predict, labels)
        test_epoch += loss.item() / len(data_loader)
    return test_epoch
```

```python

loss_fun2 = nn.CrossEntropyLoss()

def out(data_loader):
    
    epoch_max = 100
    loss_tr = []
    loss_te = []
    m = model().to(device)
    optimizer = torch.optim.AdamW(m.parameters(), lr = 0.0008)
        
    for epoch in tqdm(range(epoch_max)):
        loss_each = 0.0
        for images, labels in data_loader:
            images, labels = images.to(device).reshape(-1, 28 * 28), labels.to(device)
            predict = m(images)
            loss = loss_fun2(predict, labels)
        
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            loss_each += loss.item() / len(data_loader)
        loss_tr.append(loss_each)
        loss_te.append(test(m, test_loader_MNIST))
    return loss_tr, loss_te

loss_tr, loss_te = out(train_loader_MNIST)
```

```python
class mod_1 (nn.Module):
    def __init__(self):
        super().__init__()
        self.hide1 = nn.Linear(784, 155)
        self.hide2 = nn.Linear(155, 211)
        self.out = nn.Linear(211, 10)

    def forward(self, x):
        x = nn.functional.relu(self.hide1(x))
        x = nn.functional.relu(self.hide2(x))
        x = self.out(x)
        return x
```

```python
def explain(i, j, k):
    x = mod_1().to(device)
    for q in x.state_dict():
        q_i = i.state_dict()[q].cpu().numpy()
        q_j = j.state_dict()[q].cpu().numpy()
        x.state_dict()[q][:] = torch.Tensor((1 - k) * q_i + k * q_j)
    return x.to(device)    

mods_1 = mod_1()
mods_2 = mod_1()
mods_3 = mod_1()
mods_4 = mod_1()
```

```python

lr = [0.08, 0.01, 0.008, 0.002, 0.0008, 0.0002]
dat_tr = []
dat_te = []
dat_acc_tr = []
dat_acc_te = []
dat_sen = []
epoch_max = 50
for rate in lr:
    acc_tr = []
    acc_te = []
    loss_tr = []
    loss_te = []
    sens = []
    model = mod_1().to(device)
    loss_fun = nn.CrossEntropyLoss()
    optimizer = torch.optim.AdamW(model.parameters(), lr = rate)
        
    for epoch in tqdm(range(epoch_max)):
        epoch_sens = 0.0
        for images, labels in train_loader_MNIST:
            images, labels = images.to(device).reshape(-1, 28 * 28), labels.to(device)
            predict = model(images)
            loss = loss_fun(predict, labels)
            optimizer.zero_grad()
            grads_p = torch.autograd.grad(loss, model.parameters(), create_graph=True)[0]
            epoch_sens += grads_p.norm(2).detach().cpu().item() / len(train_loader_MNIST)

            loss.backward()
            optimizer.step()
        test_loss = test(model, test_loader_MNIST)
        train_loss = test(model, train_loader_MNIST)
        train_acc = train(model)
        test_acc = test2(model)
        
        loss_tr.append(train_loss)
        loss_te.append(test_loss)
        sens.append(epoch_sens)
        acc_tr.append(train_acc)
        acc_te.append(test_acc)
    dat_tr.append(np.mean(loss_tr))
    dat_te.append(np.mean(loss_te))
    dat_acc_tr.append(np.mean(acc_tr))
    dat_acc_te.append(np.mean(acc_te))
    dat_sen.append(np.mean(sens))
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel('Learning Rate')
ax1.set_xscale('log')
ax1.set_ylabel('Accuracy')
ax1.plot(lr, dat_acc_tr, label="Train")
ax1.plot(lr, dat_acc_te, linestyle = "dashed", label="Test")
ax1.legend()
ax2 = ax1.twinx()
ax2.set_ylabel('Sensitivity')
ax2.plot(lr, dat_sen, label="Sensitivity")
fig.tight_layout()
plt.show()
```
